```python
#converting sentence into set of strings or list of elements
words="Water is a universal solvnt"
tokens=nltk.word_tokenize(words)
print(tokens)
```

```python
#determining their pos tag
tagged=nltk.pos_tag(tokens)
tagged
```

```python
from nltk.corpus import words

# Ensure the words corpus is downloaded
nltk.download('words')
english_words = set(w.lower() for w in words.words())

# Input with a typo - using a string instead of the 'words' corpus object
input = "Water is a universal solvnt"
word_tokens = nltk.word_tokenize(input)

corrected_text = []

for word in word_tokens:
    # Check if the word is in the dictionary or is punctuation
    if word.lower() not in english_words and word.isalpha():
        # Look for suggestions with an edit distance of 1
        suggestions = []
        for w in english_words:
            if nltk.edit_distance(word.lower(), w) == 1:
                suggestions.append(w)
                break # Stop at the first match for speed

        if suggestions:
            corrected_text.append(suggestions[0])
        else:
            corrected_text.append(word)
    else:
        corrected_text.append(word)

print("Original:", input)
print("Corrected:", " ".join(corrected_text))
```
